```python
prices = pro.daily(ts_code='000006.SZ', start_date='20210101', end_date = today)
data.index = prices['trade_date']
data
```

```python
# reverse the dataframe so that the date is in ascending order
data = data.iloc[::-1]
data
```

```python
# drop securities if none-NaN values less than 250
fin_data = data.dropna(axis=1, thresh=250)
fin_data
```

```python
# fill na with bfill method
fin_data.fillna(method='bfill', inplace=True)
fin_data.describe()
```

```python
# calculate Calmar ratio
def calmar(s):
    ret=(s/s.iloc[0]-1).iloc[-1]
    drawdown=(1-s/s.cummax()).max()
    kama=ret/drawdown
    return kama

# 计算250日期间收益
data2 = fin_data[-250:]
rets = data2.apply(lambda s:(s/s[0] - 1)[-1])
drawdowns = data2.apply(lambda s:(1-s/s.cummax()).max())
calmars = data2.apply(lambda s:calmar(s))
calmar_df=pd.concat([rets.to_frame('rets'),drawdowns.to_frame('Drawndowns'),calmars.to_frame('Calmars')],axis=1)
calmar_df
```

```python
#计算多个期间序列的kama比率并排序
def stock_calmar(data,w_list=[20,60,120,250]):
    df=data.copy()
    result=pd.DataFrame()
    for w in w_list:
        temp=df[-w:].apply(lambda s:calmar(s))
        securities=temp.sort_values(ascending=False)[:20].index
        security_name = []
        for code in securities:
            security_name.append(pro.stock_basic(ts_code=code)['name'].values[0])
        result[str(w)+'日']=security_name
    return result

result=stock_calmar(fin_data)

result
```
